```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.stats as stats
from math import exp, log
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = "{:,.5f}".format
```

```python
plt.rcParams["figure.figsize"] = [3.5, 3.5]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#Hiperparámetros
rho = -0.7
mx, my = 0, 0
sx, sy = 1, 1

u, v = np.mgrid[-3:3:1000j, -3:3:1000j] #xj indica el número de pasos
x = u
y = v

f1 = 1 / (2 * np.pi * sx * sy * np.sqrt(1-rho**2))
f2 = -1/ (2 * (1 - rho**2))
f3 = ((x - mx) / sx)**2
f4 = 2 * rho * ((x - mx) / sx) * ((y - my) / sy)
f5 = ((y - my) / sy)**2
z = f1 * np.exp(f2 * (f3 - f4 + f5))

ax.plot_surface(x, y, z,cmap=plt.cm.YlGnBu_r,alpha=0.9)
ax.view_init(25, 60)
plt.title('rho = - 0.7')
plt.show()
plt.contour(x,y,z, levels = 15, cmap = "coolwarm")
```
